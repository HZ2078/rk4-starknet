```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def evaluate_CHO(t, state):
    '''
    evaluate the SHO at state (x1, x1_dot, x2, x1_dot)
    Note: this diff eq has no time dependence
    '''
    
    k1 = 17
    k2 = 10
    k3 = 15
    m1 = 1
    m2 = 2
    W = 1000
    
    [x1, x1_dot, x2, x2_dot] = state # unpack state
    
    x1_diff = x1_dot
    v1_diff = ( -k1*x1 + k2*(x2-x1) ) / m1
    x2_diff = x2_dot
    v2_diff = ( -k2*(x2-x1) + k3*(W-x2) ) / m2
    
    return np.array( [x1_diff, v1_diff, x2_diff, v2_diff]) 
```

```python
# reference: https://prappleizer.github.io/Tutorials/RK4/RK4_Tutorial.html

def rk4(t,dt,state,evaluate):
    '''
    Given a vector state at t, calculate state at t+dt
    using rk4 method
    '''
    k1 = dt * evaluate(t, state)
    k2 = dt * evaluate(t + 0.5*dt, state + 0.5*k1)
    k3 = dt * evaluate(t + 0.5*dt, state + 0.5*k2)
    k4 = dt * evaluate(t + dt, state + k3)
    
    state_delta = (1/6.)*(k1+ 2*k2 + 2*k3 + k4)
    state_t_dt = state + state_delta
    
    return state_t_dt, state_delta
```

```python
# Running a small integration
dt = 0.01
T = 5

state_0 = np.array( [150, 0, 800, 1000] )
history = [state_0]
history_delta = []
ts = [0]

nsteps = int(T/dt)
for i in range(nsteps):
    t = ts[-1]
    state_new, state_delta = rk4 (t, dt, history[-1], evaluate_CHO)
    history.append(state_new)
    history_delta.append(state_delta)
    
    t += dt
    ts.append(t)
    
history = np.array(history)
history_delta = np.array(history_delta)
ts = np.array(ts)
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,6),sharex=True)
ax[0].plot(ts,history[:,0],color='C0',lw=2,ls='-',label='m1 position',alpha=0.5)
ax[0].plot(ts,history[:,2],color='C1',lw=2,ls='-',label='m2 position',alpha=0.5)
ax[0].legend(loc="upper center")

ax[1].plot(ts,history[:,1],color='C0',lw=2,alpha=0.5,ls='-',label='m1 velocity')
ax[1].plot(ts,history[:,3],color='C1',lw=2,alpha=0.5,ls='-',label='m2 velocity')
ax[1].legend(loc="upper center")
ax[-1].set_xlabel('time');
```
