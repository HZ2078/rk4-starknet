```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reference: https://prappleizer.github.io/Tutorials/RK4/RK4_Tutorial.html

def rk4(t,dt,state,evaluate):
    '''
    Given a state of [x,xdot] at t, calculate state at t+dt
    using rk4 method
    '''
    k1 = dt * evaluate(t, state)
    k2 = dt * evaluate(t + 0.5*dt, state + 0.5*k1)
    k3 = dt * evaluate(t + 0.5*dt, state + 0.5*k2)
    k4 = dt * evaluate(t + dt, state + k3)
   
    state_t_dt = state + (1/6.)*(k1+ 2*k2 + 2*k3 + k4)
    return state_t_dt
```

```python
def evaluate_SHO(t,state,k=1):
    '''
    evaluate the SHO at state
    Note: this diff eq has no time dependence
    '''
    k = 150
    m = 10
    [x,v] = state # unpack state
    x_diff = v
    v_diff = -k/m * x
    return np.array([x_diff, v_diff])
   
```

```python
# Running a small integration

y_0 = np.array([100,0])
history = [y_0]
ts = [0]
dt = 0.01
T = 2
nsteps = int(T/dt)
for i in range(nsteps):
    t = ts[-1]
    state_new = rk4(t,dt,history[-1],evaluate_SHO)
    history.append(state_new)
    t += dt
    ts.append(t)
history = np.array(history)
ts = np.array(ts)
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,6),sharex=True)
ax[0].plot(ts,history[:,0],color='C0',lw=3,ls='-',label='Position',alpha=0.5)
ax[1].plot(ts,history[:,1],color='C1',lw=3,alpha=0.5,ls='-',label='Velocity')
ax[0].legend(loc="upper center")
ax[1].legend(loc="upper center")
ax[-1].set_xlabel('time');
```
